```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

# Lab | Getting started with Tableau

## 1 Load the dataset into a notebook.

```python
# reading the data
customer_df = pd.read_csv("we_fn_use_c_marketing_customer_value_analysis.csv")
customer_df
```

## 2. Create a barplot of the number of customers per Gender.

```python
# Count the number of customers per gender
gender_counts = customer_df['Gender'].value_counts()

# Create a barplot
plt.bar(gender_counts.index, gender_counts.values)

# Add labels and a title
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.title('Number of Customers per Gender')

# Display the plot
plt.show()
```

## 3. Create a barplot of the number of customers per EmploymentStatus and Gender.

```python
# Count the number of customers per EmploymentStatus and Gender
counts = customer_df.groupby(['EmploymentStatus', 'Gender']).size().unstack()

# Create a stacked barplot
counts.plot(kind='bar', stacked=True)

# Add labels and a title
plt.xlabel('EmploymentStatus')
plt.ylabel('Number of Customers')
plt.title('Number of Customers per EmploymentStatus and Gender')

# Display the plot
plt.show()
```
